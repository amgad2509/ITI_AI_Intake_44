## Import Pandas and numpy

```python
import pandas as pd
import numpy as np
import warnings
warnings.filterwarnings('ignore')
import re
%matplotlib inline
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("Data/Uncleaned_DS_jobs.csv")
```

## Drop duplicates

```python
df.drop_duplicates(inplace=True)
```

## Reset index and show shape

```python
df.reset_index(inplace=True)
df.shape
```

## Show column ['Company Name']

```python
df[['Company Name']]
```

## Remove numbers from Company Name

```python
df['Company Name']=df['Company Name'].str.replace('\d+','',regex=True)
df[['Company Name']]
```

## Get values count of 'Salary Estimate' column

```python
df['Salary Estimate'].value_counts()
```

## Get the lower and higher limit of the salary in new columns (Hint: use regex, split)

```python
df[['Lower_Limit', 'Higher_Limit']] = df['Salary Estimate'].str.extract(r'\$(\d+)K-\$(\d+)K')
```

```python
df
```

## get average salary

```python
#First we need to convert Data Type TO INT 
df.Lower_Limit=df.Lower_Limit.astype('int32')
df.Higher_Limit=df.Higher_Limit.astype('int32')
df.dtypes
```

```python
df['Average_Salary']=(df['Lower_Limit']+df['Higher_Limit'])//2
df
```

## Extract state code from "location" column as a new column "state"

```python
df['Location'].str.split(',')[0][1]
```

```python
df['State']=df['Location'].str.split(',').apply(lambda x: x[-1])
df
```

## Get values count of 'state' column

```python
df['State'].value_counts()
```

## replace values using the next dict
state_replacements = {

    'Texas': 'TX',

```python
state_replacements = {
    'US': 'United States',
    'TX': 'Texas',
    'CA': 'California',
    'NJ': 'New Jersey',
    'Remote': np.nan,
    'Utah': np.nan
}
df['State'] = df['State'].replace(state_replacements)
df
```

```python
df.State.value_counts()
```

## Drop na from rows and reset index

```python
df.isna().sum()
```

```python
df.dropna(inplace=True)
```

### make a new feature for each word in the following list ["python", "excel", "hadoop", "spark", "aws", "tableau", "big data"] as binary  column [0,1] if the word appeared In 'Job Description' column

```python
l = ["python", "excel", "hadoop", "spark", "aws", "tableau", "big data"]

for word in l:
    df[word] = df['Job Description'].str.lower().str.contains(word).astype(int)
df
```

## Describe the "Rating" column

```python
df['Rating'].describe().T
```

## Notice there are negative ratings? remove them

```python
Final_df=df[~(df['Rating']<0)]
Final_df
```

## Remove level_0 and index columns

```python
df.drop('index',inplace=True,axis=1)
df
```

## Plot mean salary per Sector using the mean of avg salary column you created earlier

```python
plt.figure(figsize=(20, 20))
sns.barplot(x='Sector', y='Average_Salary', data=df, palette='viridis', ci=None)
plt.xlabel('Sector')
plt.ylabel('Mean Salary')
plt.title('Mean Salary per Sector')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

## Plot count of each occurance of the skills you created earlier

```python
category_columns=['python', 'excel', 'hadoop', 'spark', 'aws', 'tableau', 'big data']
fig, axes = plt.subplots(len(category_columns), 1, figsize=(10, 10))
for i, col in enumerate(category_columns):
    sns.countplot(data=df, y=col, palette='viridis', order=df[col].value_counts().index[:2], ax=axes[i])
    axes[i].set_ylabel(col.capitalize())

plt.tight_layout()
plt.show()
```

## Show a horizontal box plot of the average salary

```python
plt.figure(figsize=(10, 10))
sns.boxplot(data=df, x='Average_Salary', orient='h', palette='viridis')
plt.xlabel('Average Salary')
plt.title('Horizontal Box Plot of Average Salary')
plt.tight_layout()
plt.show()
```
